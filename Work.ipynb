### Machine Learning

```python
import seaborn as sns
import pandas as pd 
import random
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
y=df[['Cluster']]
y
```

```python
#Now we have to only take the features acousticnes,danceability,energy,instrumentalness,liveness,valence,tempo 
# Remove two columns name is 'C' and 'D' 
# df.drop(['C', 'D'], axis = 1)  
dataset1=pd.DataFrame(df.drop(['artist','album','track_name','track_id','key','loudness','mode','speechiness','duration_ms','time_signature','Cluster'],axis=1))
dataset1.max()
dataset1
```

```python
scaler=MinMaxScaler()
scaled=pd.DataFrame(scaler.fit_transform(dataset1))
scaled.columns=['acousticness','danceability','energy','instrumentalness','liveness','valence','tempo']
scaled
```

```python
scaled.min()
```

```python
X=scaled[['acousticness','danceability','energy','instrumentalness','liveness','valence','tempo']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
X_train
X_test
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
cnf_matrix=confusion_matrix(y_test,y_pred)
cnf_matrix
```

```python
sns.heatmap(cnf_matrix,annot=True, linewidths = 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
modelRandom=RandomForestClassifier()
modelRandom.fit(X_train,y_train)
```

```python
modelRandom.score(X_test,y_test)
```

```python
y_randpred=modelRandom.predict(X_test)
y_randpred
```

```python
cnf_randmatrix=confusion_matrix(y_test,y_randpred)
cnf_randmatrix
```

```python
sns.heatmap(cnf_randmatrix,annot=True, linewidths = 1)
```
