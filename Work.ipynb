### Model Planning - Exploratory Data Analysis

```python
energy_max=df["energy"].max() #Find song with maximum energy
print(energy_max)
df.query('energy == @energy_max')
```

```python
acoustic_max=df["acousticness"].max()  #Find song with maximum acousticness
print(acoustic_max)
df.query('acousticness == @acoustic_max')
```

```python
dance_max=df["danceability"].max()  #Find song with maximum danceability
print(dance_max) 
df.query('danceability == @dance_max')
```

```python
#Plots for the Cluster 0 
fig, axes = plt.subplots(nrows=5, ncols=2,figsize=(15,30))
display=["acousticness","danceability", "loudness" ,"tempo","duration_ms",
                 "energy" ,"speechiness" ,"instrumentalness" ,
                 "liveness" ,"valence"]

yes=[0,0,0,1,1,0,1,1,2,0,2,1,3,0,3,1,4,0,4,1]
count=0
df_country = df[0:305]
for column in display:
    sns.distplot(df_country[column],ax=axes[yes[count],yes[count+1]],
                 bins=40,kde=True,color='r',norm_hist=False)
    count+=2
```
